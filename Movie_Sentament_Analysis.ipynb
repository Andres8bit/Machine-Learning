```python
def get_id(title):
  ia = IMDb()
  movie = ia.search_movie(title)
  
  return movie[0].movieID
```

```python
def get_reviews(id):
  ia = IMDb()
  reviews = ia.get_movie(id)
  ia.update(reviews,['reviews'])
  
  content = []
  for review in reviews['reviews']:
    content.append(review['content'])
  return content
```
