```python
# converting age from single feature, to multiple based on ranges:
# [0-4], [5-17], [18-29], [30-39], [40-49], [50-64], [65-74], [75-84],[85+]

fill = data_copy.iloc[:,:].copy()
ranges = pd.DataFrame(columns=['0-4', '5-17', '18-29','30-39','40-49','50-64','65-74','75-84','85'])


range_0_4_ =[0] * 8192
range_5_17_ =[0] * 8192
range_18_29_ =[0] * 8192
range_30_39_ =[0] * 8192
range_40_49_ =[0] * 8192
range_50_64_ =[0] * 8192
range_65_74_ =[0] * 8192
range_75_84_ =[0] * 8192
range_85_ = [0] * 8192
  
for i in range(0,8192):
    if data_copy['age'][i] <= 4:
      range_0_4_[i] = 1
    elif data_copy['age'][i] >= 5 and data_copy['age'][i] <= 17:
      range_5_17_[i] = 1
    elif data_copy['age'][i] >= 18 and data_copy['age'][i] <= 29:
      range_18_29_[i] = 1
    elif data_copy['age'][i] >= 30 and data_copy['age'][i] <= 39:
      range_30_39_[i] = 1
    elif data_copy['age'][i] >= 40 and data_copy['age'][i] <= 49:
      range_40_49_[i] = 1
    elif data_copy['age'][i] >= 50 and data_copy['age'][i] <= 64:
      range_50_64_[i] = 1
    elif data_copy['age'][i] >= 65 and data_copy['age'][i] <= 74:
      range_65_74_[i] = 1
    elif data_copy['age'][i] >= 75 and data_copy['age'][i] <= 84:
      range_75_84_[i] = 1
    if data_copy['age'][i] >= 85:
      range_85_[i] = 1


```

```python
data = pd.read_csv(io.BytesIO(full_data_set['covid.csv']))
data.head()
```

```python
data_copy = data.drop(columns=['id'])
```

```python
data_copy = data_copy.drop(columns=['entry_date'])
```

```python
data_copy = data_copy.drop(columns=['date_symptoms'])
```

```python
deceased  = ['N']* data_copy.shape[0]
for i in range(data_copy.shape[0]):
  if data_copy['date_died'][i] != "9999-99-99":
    deceased[i] = 'Y'


```

```python
data_copy['deceased'] = deceased
```

```python
data_copy.head()
```

```python
pat_type = ['out']* data_copy.shape[0]
for i in range(data_copy.shape[0]):
  if data_copy['patient_type'][i] == 2:
    pat_type[i] = 'in'

```

```python
data_copy['patient_type'] = pat_type
data_copy.head()
```

```python
diabetes = ['N']* data_copy.shape[0]
for i in range(data_copy.shape[0]):
  if data_copy['diabetes'][i] == 1:
    diabetes[i] = 'Y'

```

```python
data_copy['diabetes'] = diabetes
data_copy.head()
```

```python
pneumonia= ['N']* data_copy.shape[0]
for i in range(data_copy.shape[0]):
  if data_copy['pneumonia'][i] == 1:
    pneumonia[i] = 'Y'

```

```python
data_copy['pneumonia'] = pneumonia
data_copy.head()
```

```python
intubed = ['N'] * data_copy.shape[0]

for i in range(data_copy.shape[0]):
  if data_copy['intubed'][i] == 1:
    intubed[i] = 'Y'
```

```python
data_copy['intubed'] = intubed
data_copy.head()
```

```python
def to_catigorical(data,feature):
  temp = ['N'] * data.shape[0]
  for i in range(data.shape[0]):
    if data[feature][i] == 1:
      temp[i] = 'Y'
  return temp
```

```python
data_copy['pregnancy'] = to_catigorical(data_copy,'pregnancy')
data_copy.head()
```

```python
data_copy['copd'] = to_catigorical(data_copy,'copd')
data_copy.head()
```
